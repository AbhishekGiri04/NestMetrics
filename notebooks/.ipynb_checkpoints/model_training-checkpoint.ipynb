# Step 1: Import Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import joblib 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

# Step 2: Load Data

```python
df = pd.read_csv('../backend/data/Airbnb_Dataset.csv', low_memory=False)
```

# Step 3: Data Cleaning & Preprocessing

```python
# Drop duplicates
df.drop_duplicates(inplace=True)

# Drop columns if they exist
df.drop(['house_rules', 'license'], axis=1, inplace=True, errors='ignore')

# Clean price and service fee
for col in ['price', 'service fee']:
    if col in df.columns:
        df[col] = df[col].str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Rename columns safely
df.rename(columns={'price': 'price_$', 'service fee': 'service_fee_$'}, inplace=True)

# Drop missing values
df.dropna(inplace=True)

# Correct data types
df['price_$'] = df['price_$'].astype(float)
df['service_fee_$'] = df['service_fee_$'].astype(float)
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
df['Construction year'] = df['Construction year'].astype(int)

# Correct neighborhood spelling
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'

# Remove outliers
df = df[df['availability 365'] <= 500]
```

# Step 4: Exploratory Data Analysis (EDA)

```python
# Room type counts
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='room type', color='skyblue')  # use color instead of palette
plt.title('Room Type Counts')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()

# Neighborhood group counts
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='neighbourhood group', color='lightgreen')
plt.title('Listings by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Number of Listings')
plt.show()

# Average price by neighborhood
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,6))
sns.barplot(x=avg_price.index, y=avg_price.values, color='skyblue')  # replace palette with color
plt.title('Average Price by Neighborhood Group')
plt.ylabel('Price ($)')
plt.show()

# Construction year vs price
plt.figure(figsize=(12,6))
sns.lineplot(data=df.groupby('Construction year')['price_$'].mean(), color='purple')
plt.title('Average Price vs Construction Year')
plt.ylabel('Price ($)')
plt.xlabel('Construction Year')
plt.show()

# Top 10 hosts by listings
top_hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_hosts.index, y=top_hosts.values, color='orange')
plt.title('Top 10 Hosts by Listing Count')
plt.xticks(rotation=75)
plt.ylabel('Listings Count')
plt.show()

# Verified hosts review comparison
host_review = df.groupby('host_identity_verified')['review rate number'].mean()
plt.figure(figsize=(6,4))
sns.barplot(x=host_review.index, y=host_review.values, color='lightcoral')
plt.title('Average Review Rate by Host Verification Status')
plt.ylabel('Average Review')
plt.show()

# Price vs Service Fee correlation
plt.figure(figsize=(8,6))
sns.scatterplot(x='price_$', y='service_fee_$', data=df, color='teal')
plt.title('Price vs Service Fee')
plt.show()
print("Correlation (Price vs Service Fee):", df['price_$'].corr(df['service_fee_$']))

# Review by room type and neighborhood
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='neighbourhood group', y='review rate number', hue='room type')
plt.title('Average Review Rate by Room Type and Neighborhood Group')
plt.show()

# Listings count vs Availability
plt.figure(figsize=(8,6))
sns.scatterplot(x='calculated host listings count', y='availability 365', data=df, color='purple')
plt.title('Host Listing Count vs Availability')
plt.show()
print("Correlation (Listings vs Availability):", df['calculated host listings count'].corr(df['availability 365']))
```

# Step 5: Feature Engineering

```python
# Make a safe copy first 
df = df.copy()

le = LabelEncoder()
df.loc[:, 'room_type_enc'] = le.fit_transform(df['room type'])
df.loc[:, 'neighbourhood_group_enc'] = le.fit_transform(df['neighbourhood group'])
df.loc[:, 'host_verified_enc'] = le.fit_transform(df['host_identity_verified'])
```

# Step 6: Price Prediction Model

```python
X_price = df[['room_type_enc', 'neighbourhood_group_enc', 'minimum nights',
              'availability 365', 'Construction year', 'calculated host listings count', 'service_fee_$']]
y_price = df['price_$']

X_train, X_test, y_train, y_test = train_test_split(X_price, y_price, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

price_model = RandomForestRegressor(n_estimators=200, random_state=42)
price_model.fit(X_train_scaled, y_train)
y_pred_price = price_model.predict(X_test_scaled)

print("Price Prediction RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_price)))
print("Price Prediction R¬≤:", r2_score(y_test, y_pred_price))

# Feature Importance
feat_importance = pd.Series(price_model.feature_importances_, index=X_price.columns)
feat_importance.sort_values().plot(kind='barh', title='Feature Importance for Price Prediction', color='teal')
plt.show()

# Save the trained Price Prediction model
joblib.dump(price_model, "../backend/models/model.pkl")
print("Price prediction model saved at ../backend/models/model.pkl")
```

# Step 7: Review Prediction Model

```python
X_review = df[['room_type_enc', 'neighbourhood_group_enc', 'price_$', 'service_fee_$', 'host_verified_enc']]
y_review = df['review rate number']
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_review, y_review, test_size=0.2, random_state=42)

review_model = RandomForestRegressor(n_estimators=150, random_state=42)
review_model.fit(X_train_r, y_train_r)
y_pred_review = review_model.predict(X_test_r)
print("Review Prediction RMSE:", np.sqrt(mean_squared_error(y_test_r, y_pred_review)))
print("Review Prediction R¬≤:", r2_score(y_test_r, y_pred_review))
```

# Step 8: Availability Prediction Model

```python
X_avail = df[['room_type_enc', 'neighbourhood_group_enc', 'price_$', 'calculated host listings count']]
y_avail = df['availability 365']
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_avail, y_avail, test_size=0.2, random_state=42)

avail_model = RandomForestRegressor(n_estimators=150, random_state=42)
avail_model.fit(X_train_a, y_train_a)
y_pred_avail = avail_model.predict(X_test_a)
print("Availability Prediction RMSE:", np.sqrt(mean_squared_error(y_test_a, y_pred_avail)))
print("Availability Prediction R¬≤:", r2_score(y_test_a, y_pred_avail))
```

# Step 9: Save Model

```python
# Save the cleaned dataset in ../backend/models folder as Processed.csv
df.to_csv("../backend/models/Processed.csv", index=False)
print("Cleaned dataset saved as ../backend/models/Processed.csv")
```
